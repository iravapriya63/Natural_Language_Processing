```python
#question !
#part-a
import urllib.request
import os

def download_pages():
    if not os.path.exists("saved_pages"):
        os.makedirs("saved_pages")
    with open("sw-security-urls.txt", "r") as f:
        urls = f.readlines()
    for i, url in enumerate(urls):
        url = url.strip()
        response = urllib.request.urlopen(url)
        webContent = response.read()
        file_name = "saved_pages/page_{}.html".format(i)
        with open(file_name, "wb") as f:
            f.write(webContent)

download_pages()

```

```python
#part-b
from bs4 import BeautifulSoup
import json

data = []

for i in range(60):
    file_name = "saved_pages/page_{}.html".format(i)
    with open(file_name, "r", encoding="utf-8", errors='ignore') as f:
        html_content = f.read()
    soup = BeautifulSoup(html_content, 'html.parser')
    title = soup.title.string
    body_content = soup.find("div", {"id": "bodyContent"})
    json_ld = json.loads(soup.find("script", type="application/ld+json").text)
    data.append({
        "title": title,
        "main_text": body_content.text,
        "name": json_ld.get("name"),
        "url": json_ld.get("url"),
        "datePublished": json_ld.get("datePublished"),
        "headline": json_ld.get("headline")
    })
print(data)

```

```python
#part-c
text = ""
for d in data:
    text += d["main_text"] + " "

with open("main_texts.txt", "w",encoding="utf-8", errors='ignore') as f:
    f.write(text)


print(text)
```

```python
#part-d
import pandas as pd

df = pd.DataFrame(data)
df = df.set_index("title")

print(df)
```

```python
#part-d
missing_headline = df[df['headline'].isnull()]
print(missing_headline)
```

```python
#part-e
df = df.fillna("#empty#")
empty_headline_articles = df[df['headline'] == "#empty#"]
print(empty_headline_articles)

```

```python
#part-f
import sqlite3

# Connect to a local sqlite3 database
connection = sqlite3.connect("my_db.sqlite3")

# Use the to_sql() function to store the dataframe in the "articles" table
df.to_sql("articles", connection, if_exists="replace", index_label = "title")

# Close the connection
connection.close()
```
